```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load data 
completions = pd.read_csv("completions_df.csv")
completions
```

```python
emps = pd.read_csv("emps_df.csv")
emps
```

```python
Ns = range(2,6)
years = completions['year']

for N in Ns:
    y_pred = completions['completions'].rolling(window=N).mean()
    plt.plot(years, completions['completions'], label='Observed')
    plt.plot(years, y_pred, label='Predicted')
    plt.title(f"Completion MA Forecast, window={N}")
    plt.show()
```

```python
# Now that we know the window, get completion for the next 3 years
y =completions['completions']
future_ys = [list(y)[-1]]
window = 3
for i in range(window):
    y_pred = list(y.rolling(window=3).mean())
    y = list(y)
    future_ys.append(y_pred[-1])
    y.append(y_pred[-1])
    y = pd.Series(y)
```

```python
# Plot
plt.plot(completions['year'],completions['completions'])
plt.plot(completions['year'][-window-1:]+window,future_ys)
# Repeat for institution count, wages, and employment
```
